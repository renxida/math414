```python
import numpy as np
```

```python
def get_ts(h, w, a):
    def A(h, w, a):
        return np.matrix([[1-h*h*w*w*(1-a), -h*w*(a + (1-a)*(1-h*h*w*w*0.5))], [h*w, 1-h*h*w*w/2]])
    
    v0 = np.matrix([[0],[1]])

    def xf(A, step):
        return ( (A**step) * v0)

    t = [None]*50
    x = [None]*50

    for i in range(50):
        t[i] = h * i
        x[i] = xf(A(h,w,a), i).item((1,0))
    return t,x
```

```python

a=0.5
w=1
h=np.pi/10
```

```python
t, x = get_ts(h,w,a)
```

```python
m = x[0]
```

```python
import matplotlib.pyplot as plt
for a in (0, 0.5, 1):
    t,x = get_ts(h, w, a)
    plt.plot(t,x)

plt.show()
```
